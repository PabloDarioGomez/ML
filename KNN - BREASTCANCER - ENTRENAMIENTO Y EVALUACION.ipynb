```python
########## LIBRERÍAS A UTILIZAR ##########
#Se importan la librerias a utilizar
from sklearn import datasets
import numpy as np
import pandas as pd
########## PREPARAR LA DATA ##########
#Importamos los datos
dataset = pd.read_csv('cancer1.csv', encoding = "ISO-8859-1") 
dataset.shape
```

```python
########## ENTENDIMIENTO DE LA DATA ##########
#Verifico la información contenida en el dataset
print('Información en el dataset:')
print(dataset.keys())
print()
```

```python
#Seleccionamos todas las columnas, menos los targets
X = dataset.drop('target', axis=1)
#Defino los datos correspondientes a las etiquetas
y = dataset.target
```

```python
########## IMPLEMENTACIÓN DE K VECINOS MÁS CERCANOS ##########
from sklearn.model_selection import train_test_split

#Separo los datos de "train" en entrenamiento y prueba para probar los algoritmos
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

#Defino el algoritmo a utilizar
from sklearn.neighbors import KNeighborsClassifier

algoritmo = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
```

```python
#Entreno el modelo
algoritmo.fit(X_train, y_train)
#Realizo una predicción
y_pred = algoritmo.predict(X_test)
```

```python
#Verifico la matriz de Confusión
from sklearn.metrics import confusion_matrix
matriz = confusion_matrix(y_test, y_pred)
print('Matriz de Confusión:')
print(matriz)
```

```python
#Calculo la precisión del modelo
from sklearn.metrics import precision_score
precision = round(precision_score(y_test, y_pred)*100,2)
print('Precisión del modelo:')
print(precision, '%')
```
